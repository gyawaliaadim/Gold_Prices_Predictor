```python
import json
import pandas as pd
```

```python

# Load raw data
with open("data/raw_data.json", "r") as f:
    try:
        data = json.load(f)
    except json.JSONDecodeError as e:
        print(f"JSON Error: {e}")
        print(f"Line {e.lineno}, Column {e.colno}")
        raise

# Create DataFrame
df = pd.DataFrame(data)

# Split x = [year, month(0-based), day]
df[["year", "month", "day"]] = pd.DataFrame(df["x"].tolist(), index=df.index)

# Fix zero-based month
df["month"] = df["month"] + 1

# Rename price column
df["price"] = df["y"]

# Final shape
final_df = df[["year", "month", "day", "price"]]

final_df.to_csv("data/processed_Data.csv", index=False)
```
