```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("data/processed_Data.csv").dropna(how='any',axis=0).reset_index(drop=True)
df.describe()
```

```python
# 1. Prepare your data (e.g., x and y coordinates)
x_values = df.index # Creates 100 evenly-spaced points from 0 to 10
y_values = df["price"].sort_index(ascending=False)   # Calculates the sine of each x value

# 2. Create the plot
plt.figure(figsize=(8, 5)) # Optional: Adjusts the figure size
plt.plot(x_values, y_values, label='Price', color='blue', linestyle='-') # Plots the line

# 3. Add labels, a title, and a grid (optional but recommended)
plt.xlabel('Days (2019-2026)')
plt.ylabel('Gold Price')
plt.title("Line Graph of Gold Prices")
plt.grid(True)
plt.legend() # Shows the label in the plot

# 4. Display the graph
plt.show()
```
